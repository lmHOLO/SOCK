## 한국 과자 크롤링 (테스트용)

```python
# 분류 별 url 리스트를 가져오는 함수
def find_snack_links(links):
  html = urlopen(links)
  soup = BeautifulSoup(html.read(), "lxml")
  snack_list = soup.select(
    '#longlist > ul > li > a'
  )
  url_links = []
  url_links += [snack.attrs['href'] for snack in snack_list]

  return url_links
```

```python
# 한국과자
korea_url = "https://www.taquitos.net/snacks-from/Korea"
korea_url_list = find_snack_links(korea_url)

print(korea_url_list)
print(len(korea_url_list))
```
